```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('IS310.xlsx', skiprows=17, nrows=24-17)

# Extract percentiles and income from DataFrame
percentiles = df['Unnamed: 1'][1:].astype(float)  
income = df['Unnamed: 2'][1:].astype(float)  

# Plotting
plt.bar(percentiles, income)
plt.xlabel('Percentiles')
plt.ylabel('Income')
plt.title('Upper limit of families and Unrelated Individuals Income by Percentile of Pop. 1950')
plt.show()
```

```python
df = pd.read_excel('Book1.xlsx', skiprows=3)

percentiles = df['Unnamed: 5'].tolist()[1:]  #
incomes = df['Unnamed: 6'].tolist()[1:]  

# Convert data to numeric (excluding the 'percentile' label)
percentiles = [int(p) for p in percentiles]
incomes = [int(i) for i in incomes]

# Create a bar plot
plt.bar(percentiles, incomes)
plt.xlabel('Percentile')
plt.ylabel('Income')
plt.title('Upper limit of household income by percentile of pop. 2000')
plt.show()
```

```python

df = pd.read_excel('IS310.xlsx', header=1)

data = df.iloc[54:65, 1:4]

data.set_index('Unnamed: 1', inplace=True)

# Convert the data to numeric
data = data.astype(float)

# Create a line plot
ax = data.plot(kind='line', marker='o')

# Set the labels and title
plt.xlabel('Income')
plt.ylabel('Percentile')
plt.title('Percentile vs Upper Limit Income for Families and Unrelated Individuals 1950')
ax.lines[0].set_label('White')
ax.lines[1].set_label('NonWhite')
plt.legend()

# Show the plot
plt.show()
```

```python
categories = ['White', 'Black', 'Asian and Pacific Islander', 'Hispanic']
population = [44226, 30439, 55521, 33447]

# Create bar plot
plt.figure(figsize=(5, 6))
plt.bar(categories, population, color=['blue', 'orange', 'green', 'red'], width = 0.4)
plt.xlabel('Ethnicity')
plt.ylabel('Household Income')
plt.title('Household Income by Ethnicity 2000')
plt.xticks(rotation=45)  
plt.tight_layout() 
plt.show()
```
