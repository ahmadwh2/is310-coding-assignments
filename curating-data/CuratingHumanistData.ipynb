```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://humanist.kdl.kcl.ac.uk/Archives/"
# Send a GET request to the URL
response = requests.get(url)
# Check if the request was successful
    # Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.text, "html.parser")
    # Extract the desired information from the webpage
    # TODO: Add your scraping logic here
    # Print the extracted information
links = soup.find_all("a")
links = links[1:]
links
```

```python
new_url = url + links[0].get("href")
new_response = requests.get(new_url)
new_soup = BeautifulSoup(new_response.text, "html.parser")
new_links = new_soup.find_all("a")
new_links = new_links[1:-1]
new_links
```

```python
import csv

# Open the file in write mode
with open('CuratingHumanistData.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(["Volume Link", "Volume Text"])
    # Iterate over each link in new_links
    for link in new_links:
        if "humanist" in link.get("href"):
            volume_url = new_url + link.get("href")
            volume_response = requests.get(volume_url)
            volume_soup = BeautifulSoup(volume_response.text, "html.parser")
            writer.writerow([link.get_text(), volume_soup.get_text()[:10000]])
            #volume_text = volume_soup.get_text()

            # Write the first 10000 characters of the volume text to the CSV file
            #writer.writerow([volume_url, volume_text[:10000]])
file = pd.read_csv('CuratingHumanistData.csv')
file
```
